```python
import pandas as pd

# Data from the expenditure dataset
categories = ['I', 'II', 'III', 'IV', 'V', 'VI']
actual_exp = [594449, 1356817, 798273, 679869, 383976, 184063]
budget_estimates = [609014, 1051703, 1011886, 809701, 381305, 220843]
revised_estimates = [700541, 1195078, 1021631, 813791, 415351, 211065]
budget_estimates_2022_2023 = [692214, 1181084, 1132813, 940651, 442780, 192108]

# Create a DataFrame from the data
data = pd.DataFrame({
    'Category': categories,
    'Actual': actual_exp,
    'Budget Estimates': budget_estimates,
    'Revised Estimates': revised_estimates,
    'Budget Estimates 2022-2023': budget_estimates_2022_2023
})

# Calculate the total expenditure for each fiscal year
total_actual_exp = sum(actual_exp)
total_budget_estimates = sum(budget_estimates)
total_revised_estimates = sum(revised_estimates)
total_budget_estimates_2022_2023 = sum(budget_estimates_2022_2023)

# Calculate the percentage change in expenditure from previous fiscal year
percentage_change_budget_estimates = ((total_budget_estimates - total_actual_exp) / total_actual_exp) * 100
percentage_change_revised_estimates = ((total_revised_estimates - total_budget_estimates) / total_budget_estimates) * 100
percentage_change_budget_estimates_2022_2023 = ((total_budget_estimates_2022_2023 - total_revised_estimates) / total_revised_estimates) * 100

# Display the inferences
print("Inferences from the expenditure data:")
print("---------------------------------------")
print("Total actual expenditure: {:.2f}".format(total_actual_exp))
print("Total budget estimates: {:.2f}".format(total_budget_estimates))
print("Total revised estimates: {:.2f}".format(total_revised_estimates))
print("Total budget estimates 2022-2023: {:.2f}".format(total_budget_estimates_2022_2023))
print("\nPercentage change in budget estimates from actual: {:.2f}%".format(percentage_change_budget_estimates))
print("Percentage change in revised estimates from budget estimates: {:.2f}%".format(percentage_change_revised_estimates))
print("Percentage change in budget estimates 2022-2023 from revised estimates: {:.2f}%".format(percentage_change_budget_estimates_2022_2023))
```
